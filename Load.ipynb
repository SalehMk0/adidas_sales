```python
import pandas as pd
```

```python
df = pd.read_csv('adidas.csv')
```

### Loading Dimensions
<li>cities</li>
<li>states</li>
<li>regions</li>
<li>products</li>
<li>retailers</li>
<li>sale_methods</li>

```python
regions = df['Region'].unique()
```

```python
regions_df = pd.DataFrame({
    'region_id': [i for i in range(len(regions))],
    'regions': regions
})
```

```python
regions_df
```

```python
states = df['State'].unique()
```

```python
states_df = pd.DataFrame({
    'state_id': [i for i in range(len(states))],
    'states'  : states
})
```

```python
cities = df['City'].unique()
```

```python
cities_df = pd.DataFrame({
    'city_id': [i for i in range(len(cities))],
    'city'  : cities
})
```

```python
products = df['Product'].unique()
```

```python
products_df = pd.DataFrame({
    'product_id': [i for i in range(len(products))],
    'product'  : products
})
```

```python
retailers = df['Retailer'].unique()
```

```python
retailers_df = pd.DataFrame({
    'retailer_id': [i for i in range(len(retailers))],
    'retailer'  : retailers
})
```

```python
methods = df['sales_method'].unique()
```

```python
methods_df = pd.DataFrame({
    'method_id' : [i for i in range(len(methods))],
    'method' : methods

})
```

```python
state_region_df = df[['State', 'Region']].drop_duplicates().reset_index(drop=True)

state_region_df['State_Id'] = state_region_df.index
state_region_df['Region_Id'] = state_region_df.groupby('Region').ngroup()

state_region_df = state_region_df[['State_Id', 'State', 'Region_Id', 'Region']]

state_to_region = dict(zip(state_region_df['State'], state_region_df['Region']))



states_df = state_region_df[['State_Id', 'State', 'Region_Id']]
regions_df = state_region_df[['Region_Id', 'Region']].drop_duplicates().reset_index(drop=True)
```

```python
states_df
```

```python
cities_df = df[['City', 'State']].drop_duplicates().reset_index(drop=True)

cities_df = cities_df.merge(state_region_df[['State', 'State_Id']], on='State', how='left')

cities_df['City_Id'] = cities_df.index

cities_df = cities_df[['City_Id', 'City', 'State_Id']]
```

```python
cities_df
```

```python
regions_df.to_csv('csv/d_regions.csv', index=False)
```

```python
states_df.to_csv('csv/d_states.csv', index=False)
```

```python
cities_df.to_csv('csv/d_cities.csv', index=False)
```

```python
f_adidas = df.merge(cities_df,
                    on='City',
                    how='inner'
                   )
```

```python
f_adidas
```

```python
methods_df = methods_df.rename(columns={'method' : 'sales_method'})
```

```python
f_adidas = f_adidas.merge(methods_df,
                         on='sales_method',
                         how='inner')
```

```python
f_adidas
```

```python
f_adidas = f_adidas.rename(columns={'Retailer':'retailer'})
```

```python
f_adidas = f_adidas.merge(retailers_df,
                         on='retailer',
                         how='inner')
```

```python
f_adidas
```

```python
f_adidas = f_adidas.rename(columns={'Product':'product'})
```

```python
f_adidas = f_adidas.merge(products_df,
                         on='product',
                         how='inner')
```

```python
f_adidas
```

```python
f_adidas.drop(columns=['retailer','Region','State','City','product','sales_method','State_Id'], inplace=True)
```

```python
f_adidas['sale_id'] = [i for i in range(0,len(f_adidas))]
```

```python
f_adidas
```

```python
f_adidas.to_csv('csv/adidas.csv', index=False)
```
